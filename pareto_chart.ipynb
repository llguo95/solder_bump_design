```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("datasets\\solder_ball_conc.csv", index_col=0)
df
```

```python
names = [
    "d_pad",
    "t_pad",
    "d_us",
    "d_rep1",
    "t_ubm",
    "del_d",
    "h_ball",
]

X_bounds = np.array(
    [
        [150.0, 180.0],
        [15.0, 30.0],
        [180.0, 220.0],
        [30.0, 160.0],
        [5.0, 15.0],
        [10.0, 40.0],
        [130.0, 165.0],
    ]
)

df_X_scaled = (df[names] - X_bounds[:, 0]) / (X_bounds[:, 1] - X_bounds[:, 0])
df_y_scaled = (df["max_conc"] - df["max_conc"].mean()) / df["max_conc"].std()
```
